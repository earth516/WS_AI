```python
import pandas as pd
data=pd.read_csv('/content/서울시 상권분석서비스(집객시설-자치구).csv',encoding='cp949')
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 집객시설 관련 열만 선택
columns_of_interest = [
    '집객시설_수', '관공서_수', '은행_수', '종합병원_수',
    '일반_병원_수', '약국_수', '유치원_수', '대학교_수',
    '백화점_수', '슈퍼마켓_수', '극장_수', '숙박_시설_수',
    '공항_수', '철도_역_수', '버스_터미널_수', '지하철_역_수', '버스_정거장_수'
]

# 선택한 열로 새로운 데이터프레임 생성
selected_data = data[columns_of_interest]

# 결측치 처리 (결측치를 0으로 대체)
selected_data = selected_data.fillna(0)

# 상관계수 계산
correlation_matrix = selected_data.corr()

# 상관계수 히트맵 시각화
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Facilities")
plt.show()
```

극장, 병원, 은행이 백화점과 상관계수가 높음

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 데이터 준비
columns_of_interest = ['백화점_수', '극장_수', '은행_수']
selected_data = data[columns_of_interest].fillna(0)  # 결측값 0으로 대체

# 데이터 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_data)

# K-평균 클러스터링 수행 (k=4 설정)
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# 결과를 데이터프레임에 추가
data['Cluster'] = clusters

# 결과 시각화 (2D로 극장 수와 은행 수 비교)
plt.figure(figsize=(10, 6))
plt.scatter(selected_data['극장_수'], selected_data['은행_수'], c=clusters, cmap='viridis', s=50, alpha=0.7)
plt.scatter(
    kmeans.cluster_centers_[:, 1],  # 극장 수의 클러스터 중심
    kmeans.cluster_centers_[:, 2],  # 은행 수의 클러스터 중심
    c='red', s=200, marker='X', label='Centroids'
)
plt.xlabel('극장 수')
plt.ylabel('은행 수')
plt.title('K-Means Clustering: 극장 수 vs 은행 수')
plt.legend()
plt.grid(True)
plt.show()
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Elbow Plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 데이터 준비
columns_of_interest = ['자치구_코드_명', '백화점_수', '극장_수', '은행_수']
selected_data = data[columns_of_interest].fillna(0)  # 결측치 0으로 대체

# 데이터 정규화
scaler = StandardScaler()
scaled_features = scaler.fit_transform(selected_data[['백화점_수', '극장_수', '은행_수']])

# K-평균 클러스터링 수행 (k=3)
kmeans = KMeans(n_clusters=3, random_state=42)
selected_data['Cluster'] = kmeans.fit_predict(scaled_features)

# 클러스터 중심 분석
cluster_centers = pd.DataFrame(
    scaler.inverse_transform(kmeans.cluster_centers_),
    columns=['백화점_수', '극장_수', '은행_수']
)
cluster_centers.index.name = 'Cluster'

# 번화가 클러스터 찾기 (백화점, 극장, 은행 수 합이 가장 큰 클러스터)
cluster_centers['Total'] = cluster_centers.sum(axis=1)
busiest_cluster = cluster_centers['Total'].idxmax()

# 번화가 클러스터에 속한 자치구 출력
busiest_districts = selected_data[selected_data['Cluster'] == busiest_cluster]

# 결과 출력
print(f"가장 번화한 클러스터: Cluster {busiest_cluster}")
print("\n클러스터 중심:")
print(cluster_centers)
print("\n번화가 클러스터에 속한 자치구:")
print(busiest_districts[['자치구_코드_명', '백화점_수', '극장_수', '은행_수']])
```
